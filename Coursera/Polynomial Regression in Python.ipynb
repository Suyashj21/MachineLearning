```python
import numpy as np
x = np.array([0,1,2,3,4,5])
y = np.array([0,0.8,0.9,0.1,-0.8,-1])
```

```python
from scipy import polyfit
p1 = polyfit(x,y,1)
print(p1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x,y,'o')
```

```python
plt.plot(x,y,'o')
plt.plot(x,np.polyval(p1,x),'r-')
```

```python
p2 = polyfit(x,y,2)
p3 = polyfit(x,y,3)
print("p1: " ,p1," and p2: ",p2," and p3: " ,p3)
```

```python
plt.plot(x,y,'o')
plt.plot(x,np.polyval(p1,x),'r-')
plt.plot(x,np.polyval(p2,x),'b--')
plt.plot(x,np.polyval(p3,x),'m:')
```
