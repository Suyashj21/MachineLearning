### Ordinal Number Incoding
> where we need to take care of Rank or order of the categories
> Grade or Experience or Education etc.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'date':pd.date_range('2020-01-23', periods=17, freq='D')})
df.head()
```

```python
df['name'] = pd.to_datetime(df['date']).dt.day_name()
df.head()
```

```python
#encode = {k:i for i,k in enumerate(df.name.unique(),1)}
encode = {
'Monday': 1,
'Tuesday': 2,
'Wednesday': 3,
'Thursday': 4,
'Friday': 5,
'Saturday': 6,
'Sunday': 7,
}
```

```python
df['encode'] = df.name.map(encode)
df.head()
```

#### Count of Frequency Encoding

```python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
df.head()
```

```python
df = df[[1,3,5,6,7,8,9,13]]
df.head()
```

```python
cols = ['Employment','Degree','Status','Designation','Family_job','Race','Sex','Country']
df.columns = cols
df.head()
```

```python
for feature in df.columns:
    print(feature,':',len(df[feature].unique()))
```

```python
country_count = df['Country'].value_counts().to_dict()
country_count
```

```python
df['Country'] = df['Country'].map(country_count)
df.head()
```

#### Target Guided Oridinal Encoding
1. Ordering the labels according to the target
2. Replace the labels by the join probability of being 1 or 0

```python
import numpy as np
import pandas as pd

df = pd.read_csv('datasets/titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace=True)
df.head()
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
labels = {k:i for i,k in enumerate(df.groupby('Cabin')['Survived'].mean().sort_values().index, 0)}
labels
```

```python
df['Cabin_ordinal_labels'] = df['Cabin'].map(labels)
df.head()
```

### Mean Encoding

```python
label_1 = df.groupby(['Cabin'])['Survived'].mean().to_dict()
label_1
```

```python
df['Mean_ordinal_encoding'] = df['Cabin'].map(label_1)
df.head()
```
