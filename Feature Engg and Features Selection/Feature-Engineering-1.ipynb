```python
df = pd.read_csv("datasets/mercedes.csv",usecols=['X0','X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
cate = list(df['X0'].value_counts()[:10].index)
cate
```

```python
for label in cate:
    df[label] = np.where(df['X0']==label,1,0)
df.head()
```

```python
cate.append('X0')
df[cate].head(10)
```

#### Ordinal Number Encoding

```python
dl = pd.DataFrame({'date' : pd.date_range('2020/03/01',periods=13,freq='D')})
dl
```

```python
dl['weekday'] = dl['date'].dt.day_name()
dl['weekday']
```

```python
map_label = {k:i for i,k in enumerate(dl['weekday'].unique())}
map_label
```

```python
dl['weekday_new'] = dl['weekday'].map(map_label)
dl
```
